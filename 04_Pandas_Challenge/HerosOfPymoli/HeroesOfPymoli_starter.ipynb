```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count = purchase_data["SN"].nunique()
player_total = pd.DataFrame({"Total" : [player_count]})
player_total
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = purchase_data["Item Name"].nunique()
#unique_item
purchase = purchase_data["Price"].count()
#purchase
revenue = purchase_data["Price"].sum()
#revenue
average_price = round(revenue/purchase, 2)
#average_price
price_summary = pd.DataFrame({
    "Number of unique items" : [unique_item],
    "Average Price": [average_price],
    "Number of purchases" : [purchase],
    "Total Revenue" : [revenue]})
price_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#total = purchase_data["Gender"].count()
#total
#code for getting a total count of players, dropping the duplicates
player = purchase_data.drop_duplicates(["SN"])
df_player = pd.DataFrame(player)
#getting a count for genders
gender = df_player["Gender"].value_counts()
total = df_player["Gender"].count()
gender_percent = gender/total
#making a table and giving the names and table content
df_gender = pd.DataFrame({
    "Gender" : ["Male", "Female", "Others"],
    "Total Count" : gender,
    "Percentage of PLayers" : gender_percent,})
#df_gender
#formatting
df_gender["Percentage of PLayers"] = df_gender["Percentage of PLayers"].map(('{:.2%}').format)
#making a index as gender
df_gender = df_gender.set_index("Gender")
df_gender
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#code for counting how many purchase each gender had
male_purchases = purchase_data[purchase_data["Gender"] == "Male"]["Price"].count()
female_purchases = purchase_data[purchase_data["Gender"] == "Female"]["Price"].count()
neither_purchases = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]["Price"].count()

#code for getting the average of each gender and the calculation
male_purchase_average = purchase_data[purchase_data["Gender"] == "Male"]["Price"].mean()
female_purchases_average = purchase_data[purchase_data["Gender"] == "Female"]["Price"].mean()
neither_purchases_average = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]["Price"].mean()

#code for getting the toal price of each gender
male_purchases_total = purchase_data[purchase_data["Gender"] == "Male"]["Price"].sum()
female_purchases_total = purchase_data[purchase_data["Gender"] == "Female"]["Price"].sum()
neither_purchases_total = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]["Price"].sum()

#code for getting players
persons = purchase_data.groupby(["SN"])
persons_sum = persons.sum()
persons_gender = persons["Gender"].first()
per_person = pd.concat([persons_sum, persons_gender], 1).reset_index()

#code for getting the average purchase made per person
male_per_person = per_person[per_person["Gender"] == "Male"]["Price"].mean()
female_per_person = per_person[per_person["Gender"] == "Female"]["Price"].mean()
neither_per_person = per_person[per_person["Gender"] == "Other / Non-Disclosed"]["Price"].mean()

#code for making the table and setting up each list
gender_purchase_analysis = pd.DataFrame({
    "Gender": ["Male", "Female", "Other"],
    "Purchase Count": [male_purchases, female_purchases, neither_purchases],
    "Average Purchase Price": [male_purchase_average, female_purchases_average, neither_purchases_average],
    "Total Purchase Value": [male_purchases_total, female_purchases_total, neither_purchases_total],
    "Average Purchase Per Person": [male_per_person, female_per_person, neither_per_person]})

#formatting
gender_purchase_analysis = gender_purchase_analysis.set_index("Gender")
gender_purchase_analysis["Average Purchase Price"] = gender_purchase_analysis["Average Purchase Price"].map(('${0:,.2f}').format)
gender_purchase_analysis["Total Purchase Value"] = gender_purchase_analysis["Total Purchase Value"].map(('${0:,.2f}').format)
gender_purchase_analysis["Average Purchase Per Person"] = gender_purchase_analysis["Average Purchase Per Person"].map(('${0:,.2f}').format)

gender_purchase_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
person_df = pd.DataFrame(purchase_data.groupby("SN").first())
#setting up the values for the age groups
bins = [0,9,14,19,24,29,34,39,99]
#label and sortting
age_label = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_demographics = pd.DataFrame({"Age Range" : ["<10" , "10-14" , "15-19" , "20-24" , "25-29" , "30-34" , "35-39" , "40+"]})
#making age range the index
age_demographics = age_demographics.set_index("Age Range")
age = pd.cut(person_df["Age"], bins, labels = age_label)
age_demographics["Total Count"] = age.value_counts()
age_total = age_demographics ["Total Count"]
number_ages = person_df["Age"].count()
#code for calculating the percentage of players
age_demographics["Percentage of Players"] = age_total/number_ages
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map(("{:.2%}").format)
age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#making labels for each groups
age_1 = purchase_data
bins = [0,9,14,19,24,29,34,39,100]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#making the table
age_1["Age Range"] = pd.cut(purchase_data["Age"], bins, labels = age_labels)
#getting the total item, average and total amount spend on each
age_purchase = age_1[["Age Range", "Price"]].groupby("Age Range").count()
age_purchase["Average Purchase Price"] = age_1[["Age Range", "Price"]].groupby("Age Range").mean()
age_purchase["Total Purchase Value"] = age_1[["Age Range", "Price"]].groupby("Age Range").sum()
#grouping them by user
people_age = age_1.groupby("SN")

age_sums = people_age.sum()
age_ranges = people_age["Age Range"].first()
age_per = pd.concat([age_sums, age_ranges], 1).reset_index()
#formatting 
age_purchase["Average Total Purchase Per Person"] = age_per[["Age Range", "Price"]].groupby("Age Range").mean()
age_purchase = age_purchase.rename(columns={"Price": "Purchase Count"})
age_purchase["Average Purchase Price"] = age_purchase["Average Purchase Price"].map(('${0:,.2f}').format)
age_purchase["Total Purchase Value"] = age_purchase["Total Purchase Value"].map(('${0:,.2f}').format)
age_purchase["Average Total Purchase Per Person"] = age_purchase["Average Total Purchase Per Person"].map(('${0:,.2f}').format)
age_purchase
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#counting how many times a name have showed
top_spenders = purchase_data[["SN", "Price"]].groupby("SN").count()
#average purchase price
top_spenders["Average Purchase Price"] = purchase_data[["SN", "Price"]].groupby("SN").mean()
#average value
top_spenders["Total Purchase Value"] = purchase_data[["SN", "Price"]].groupby("SN").sum()
#Rename
top_spenders = top_spenders.rename(columns={"Price": "Purchase Count"})
top_spenders = top_spenders.sort_values("Purchase Count", ascending=False)
#formatting
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map(('${0:,.2f}').format)
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map(('${0:,.2f}').format)

top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#getting all the item ids
popular = purchase_data[["Item ID", "Item Name"]].groupby("Item ID").first()
#sorting it by item id and getting the total
popular["Purchase Count"] = purchase_data[["Price", "Item ID"]].groupby("Item ID").count()
#getting the item price
popular["Item Price"] = purchase_data[["Price", "Item ID"]].groupby("Item ID").first()
#getting the revenue
popular["Total Revenue"] = purchase_data[["Price", "Item ID"]].groupby("Item ID").sum()
#getting the total
popular = popular.groupby(["Item ID", "Item Name"]).sum()
popular_df = popular.sort_values(["Purchase Count"], ascending=False)
#formatting
popular_df["Item Price"] = popular_df["Item Price"].map(('${0:,.2f}').format)
popular_df["Total Revenue"] = popular_df["Total Revenue"].map(('${0:,.2f}').format)
popular_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#finding out the item sorted by price in a decending order.
profitable = pd.DataFrame(popular.sort_values(["Total Revenue"], ascending=False))
#formatting
profitable["Item Price"] = profitable["Item Price"].map(('${0:,.2f}').format)
profitable["Total Revenue"] = profitable["Total Revenue"].map(('${0:,.2f}').format)
profitable.head()
```
